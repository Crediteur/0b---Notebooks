```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
import tensorflow as tf
from tensorflow.keras.datasets import cifar10

(x_train, y_train), (x_test, y_test) = cifar10.load_data()
print('x_train shape:', x_train.shape)
print('y_train shape:', y_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
```

```python
# mapping of label to names
target_names={
    0:"airplane",
    1:"automobile",
    2:"bird",
    3:"cat",
    4:"deer",
    5:"dog",
    6:"frog",
    7:"horse",
    8:"ship",
    9:"truck"}
   
```

```python
# plot single image sample from data set
def show_sample(sample=x_train[0], label=6):
    data = sample

    # set plot parameters
    plt.figure(figsize=(2,2))
    plt.axis('off')
    plt.title(target_names[label])

    # Use imshow to plot pixels
    plt.imshow(data)

    # Display the plot
    plt.show()


show_sample(x_train[1], y_train[1][0])
```

```python
def show_multi(samples=x_train[0:10], labels=[6,9,9,4,1,1,2,7,8,3]):
    length = len(samples)

    # create horizontal figure with subplots
    fig, axes = plt.subplots(1, length, figsize=(8,8))
    axes = axes.flatten()

    # create subplot for each image data
    for i in range(length):
        data = samples[i]
        label = labels[i]

        # plot parameters
        plt.subplot(1, length, i+1)    
        plt.axis('off')
        
        # decrease readibility of long words
        if target_names[labels[i]] == 'automobile': fsize=7 
        else: fsize=10
        
        plt.title(target_names[labels[i]], fontsize=fsize)
        plt.imshow(samples[i])
    
    # display sample count and show plot    
    print(f'{length} sample count')
    plt.show()

show_multi(x_train[31:40], [y[0] for y in y_train[31:40]])
```
